```python
_X = np.random.rand(306, 2)
_Y = np.random.rand(306)
# _X = df[['TV', 'radio']].values
# _Y = df['sales'].values
```

```python
def update_weights(W, X, Y, lr):
    d_w1 = 0
    d_w2 = 0
    N = len(Y)
    
    for i in range(N):
        d_w1 += -X[i, 0] * (Y[i] - ((W[0] * -X[i, 0]) + (W[1] * -X[i, 1])))
        d_w2 += -X[i, 1] * (Y[i] - ((W[0] * -X[i, 0]) + (W[1] * -X[i, 1])))
        
    W[0] -= lr * (d_w1/float(N))
    W[1] -= lr * (d_w2/float(N))
    return W

lr = 0.001
W = np.zeros(2)

for _ in range(1000):
    W = update_weights(W, _X, _Y, lr)
    
#     if _ % 100 == 0:
#         print("--------------")
#         print(f"W1: {W[0]}")
#         print(f"W2: {W[1]}")
#         print("--------------")
        
    
```
